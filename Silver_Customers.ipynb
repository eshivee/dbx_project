```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.read.format('parquet').load('abfss://bronze@etedbxcrop.dfs.core.windows.net/customers')
```

```python
df = df.drop('_rescued_data')
df.display()
```

```python
df = df.withColumn('domains', split(col('email'), '@')[1])
df.display()
```

```python
df.groupBy('domains').agg(count('*').alias('tot_num')).orderBy(col('tot_num').desc()).display()
```

```python
df.filter(col('domains') == 'gmail.com').display()
```

```python
df = df.withColumn('fullname', concat('first_name', lit(' '),  'last_name'))\
    .drop('last_name', 'first_name')

df.display()
```

```python
df.write.format('delta').mode('append').save('abfss://silver@etedbxcrop.dfs.core.windows.net/customers')
```
